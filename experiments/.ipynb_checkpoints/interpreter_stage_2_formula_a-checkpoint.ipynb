```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
```

```python
X_full = pd.read_csv('./datasets/speaker_formula_c.csv')
X_full
```

```python
data = pd.read_csv("./datasets/accent-mfcc-data-1.csv")
# data = pd.read_csv("https://raw.githubusercontent.com/SahilSinhaLpu/Machine-Learning/master/Datasets/SomvervilleHappines.csv")
data.isnull().sum().sum()
data
```

```python
y = data['language']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_full, y, test_size = 0.2, random_state = 42)
```

### KNN

```python
reg_knn = KNeighborsClassifier()
reg_knn.fit(X_train, y_train)
# y_pred = reg_knn.predict(X_test)
# print(metrics.classification_report(y_test, y_pred))
print("####### CROSS VAL SCORE #######")
print(cross_val_score(reg_knn, X_full, y, cv=10, scoring='accuracy').mean())
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
reg_log = LogisticRegression()
reg_log.fit(X_train, y_train)
# y_pred = reg_log.predict(X_test)
# print(metrics.classification_report(y_test, y_pred))
# print("####### CROSS VAL SCORE #######")
print(cross_val_score(reg_log, X_full, y, cv=10, scoring='accuracy').mean())
```

### SVM

```python
from sklearn.svm import SVC
reg_svc = SVC(kernel='linear')
reg_svc.fit(X_train, y_train)
# y_pred = reg_svc.predict(X_test)
# print(metrics.classification_report(y_test, y_pred))
print("####### CROSS VAL SCORE #######")
print(cross_val_score(reg_svc, X_full, y, cv=10, scoring='accuracy').mean())
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
reg_rf = GaussianNB()
reg_rf.fit(X_train, y_train)
# y_pred = reg_rf.predict(X_test)
# print(metrics.classification_report(y_test, y_pred))
print("####### CROSS VAL SCORE #######")
print(cross_val_score(reg_rf, X_full, y, cv=10, scoring='accuracy').mean())
```

### Random Forest

```python
rfc = RandomForestClassifier(random_state=0)
# fit the model
rfc.fit(X_train, y_train)
print(cross_val_score(rfc, X_full, y, cv=10, scoring='accuracy').mean())
```

### Multilayer Perceptron

```python
clf = MLPClassifier(random_state=1)
clf.fit(X_train, y_train)
print(cross_val_score(clf, X_full, y, cv=10, scoring='accuracy').mean())
```
