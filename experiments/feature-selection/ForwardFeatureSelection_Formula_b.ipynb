```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures


from sklearn.naive_bayes import GaussianNB
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.metrics import confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
X = pd.read_csv('../datasets/wine_formula_b.csv')
X
```

```python
df = pd.read_csv('../datasets/winequality-white.csv', sep=';')
# df = pd.read_csv("https://raw.githubusercontent.com/SahilSinhaLpu/Machine-Learning/master/Datasets/SomvervilleHappines.csv")
df

### Speaker DS
# Y = df['language']

### Algerian DS
# Y = df['Classes']

### Banknote DS
# Y = df['Class']

### User knowledge DS
# Y = df['UNS']

## Wine quality DS
Y = df['quality']

### Somerville DS
# Y = df['D']

### Pima Indians Diabetes DS
# Y = df['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.15)
```

### KNN

```python
%%time
plain_sfs = SFS(KNeighborsClassifier(), 
          k_features=(1, 10), 
          forward=True, 
          floating=False,
          scoring='accuracy', 
          cv=10)

plain_sfs.fit(X, Y)
```

```python
selected_features = X.columns[list(plain_sfs.k_feature_idx_)]
x_t = X_train[selected_features]
clf = KNeighborsClassifier()
clf.fit(x_t, y_train)
print(cross_val_score(clf, x_t, y_train, cv=10).mean())
```

### Logistic Regression

```python
plain_sfs_lr = SFS(LogisticRegression(), 
          k_features=(1, 10), 
          forward=True, 
          floating=False, 
          cv=10)

plain_sfs_lr.fit(X, Y)

selected_features = X.columns[list(plain_sfs_lr.k_feature_idx_)]
x_t = X_train[selected_features]
clf = LogisticRegression()
clf.fit(x_t, y_train)

print(cross_val_score(clf, x_t, y_train, cv=10).mean())
```

### Naive Bayes

```python
naive_bayes = GaussianNB()
plain_sfs_nb = SFS(naive_bayes, 
          k_features=(1, 10), 
          forward=True, 
          floating=False, 
          cv=10)

plain_sfs_nb.fit(X, Y) 
```

```python
selected_features = X.columns[list(plain_sfs_nb.k_feature_idx_)]
x_t = X_train[selected_features]
clf = GaussianNB()
clf.fit(x_t, y_train)
print(cross_val_score(clf, x_t, y_train, cv=10).mean())
```

### SVM

```python
plain_sfs_svm = SFS(svm.SVC(kernel='linear'), 
          k_features=(1, 10), 
          forward=True, 
          floating=False, 
          cv=10)
plain_sfs_svm.fit(X, Y)
```

```python
selected_features = X.columns[list(plain_sfs_svm.k_feature_idx_)]
x_t = X_train[selected_features]
clf = svm.SVC(kernel='linear')
clf.fit(x_t, y_train)
print(cross_val_score(clf, x_t, y_train, cv=10).mean())
```
